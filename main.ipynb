```python
import numpy as np 
import matplotlib.pyplot as plt
import networkx as nx

from models.model_output import SeirModelOutput
from models.seir_compartmental_model import SeirModel
from models.seir_network_model import SeirNetworkModel
from models.seir_meanfield_model import SeirMeanFieldModel
from models.seir_ebcm_model import SeirEbcmModel
```

### Compartmental SEIR

```python
seir_compartmental_model = SeirModel(population=10**6)
seir_compartmental_output = SeirModelOutput(*seir_compartmental_model.simulate(beta=1))

fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(seir_compartmental_output.daily_incidence, lw=2, color='RoyalBlue')
ax.set_title('SEIR compartmental model')
ax.set_xlabel('Days')
ax.set_ylabel('Incidence, cases')
ax.grid()
```

### SEIR Network

```python
N=10**5
G=nx.barabasi_albert_graph(N, 5) #create a barabasi-albert graph
```
